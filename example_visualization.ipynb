```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Publication-quality setup
sns.set_theme(style="whitegrid")       # clean theme
sns.set_context("talk", font_scale=1.1)  # larger fonts for publication
plt.rcParams["figure.dpi"] = 300        # high resolution
```

```python
# Example: assume df is your DataFrame
# Replace 'BMI' with your numeric feature
df = pd.read_csv("obesity.csv")
feature = "Age"
```

```python
plt.figure(figsize=(7, 5))
ax = sns.histplot(df[feature], bins=30, kde=True, color="#1f77b4", edgecolor="black", alpha=0.7)

# Titles and labels
ax.set_title(f"Distribution of {feature}", fontsize=16, fontweight="bold", pad=15)
ax.set_xlabel(feature, fontsize=13)
ax.set_ylabel("Count", fontsize=13)

# Aesthetic tweaks
sns.despine()  # remove top/right spines
plt.tight_layout()
plt.show()
```

```python
mean_val = df[feature].mean()
median_val = df[feature].median()

plt.figure(figsize=(7, 5))
sns.histplot(df[feature], bins=30, kde=True, color="#2ca02c", edgecolor="black", alpha=0.7)
plt.axvline(mean_val, color="red", linestyle="--", linewidth=2, label=f"Mean: {mean_val:.2f}")
plt.axvline(median_val, color="blue", linestyle="-.", linewidth=2, label=f"Median: {median_val:.2f}")

plt.title(f"Distribution of {feature}", fontsize=16, fontweight="bold")
plt.xlabel(feature, fontsize=13)
plt.ylabel("Frequency", fontsize=13)
plt.legend()
sns.despine()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.violinplot(x="Gender", y=feature, hue="Gender",
               data=df, palette="Set2", inner="quartile", legend=False)
plt.title(f"{feature} Distribution by Obesity Level", fontsize=16, fontweight="bold")
plt.xlabel("Obesity Level", fontsize=13)
plt.ylabel(feature, fontsize=13)
sns.despine()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Example categorical feature (e.g., "Obesity_Level")
feature = "Gender"

# Example data
df = pd.read_csv("obesity.csv")

# Count occurrences
counts = df[feature].value_counts().sort_index()  # or .sort_values() for order by frequency
```

```python
sns.set_theme(style="whitegrid")
sns.set_context("talk", font_scale=1.1)
plt.rcParams.update({
    "figure.dpi": 300,              # High-resolution output
    "axes.spines.top": False,
    "axes.spines.right": False,
    "axes.labelsize": 13,
    "axes.titlesize": 16,
    "axes.titleweight": "bold",
    "font.family": "serif",
    "savefig.bbox": "tight"
})
```

```python
plt.figure(figsize=(7,5))

ax = sns.barplot(
    x=counts.index,
    y=counts.values,
    hue=counts.index, 
    palette="Set2",
    edgecolor="black"
)

# Annotate counts on top of bars
for i, val in enumerate(counts.values):
    ax.text(i, val + counts.max() * 0.02, f"{val}", ha="center", fontsize=11)

# Titles and labels
ax.set_title(f"Distribution of {feature}", pad=15)
ax.set_xlabel(feature)
ax.set_ylabel("Count")

sns.despine()
plt.tight_layout()
plt.show()
```

```python
plot_df = df[feature].value_counts().reset_index()
plot_df.columns = [feature, "Count"]

ax = sns.barplot(
    data=plot_df,
    x=feature,
    y="Count",
    hue=feature,
    palette="Set2",
    legend=False,
    edgecolor="black"
)
```

```python
for i, val in enumerate(plot_df["Count"]):
    ax.text(i, val + plot_df["Count"].max() * 0.02, f"{val}", 
            ha="center", fontsize=11)

ax.set_title(f"Distribution of {feature}", fontsize=16, fontweight="bold", pad=15)
ax.set_xlabel(feature, fontsize=13)
ax.set_ylabel("Count", fontsize=13)

sns.despine()
plt.tight_layout()
plt.show()
```
